# Count of loan applicants who have good income  

```python
import math

ApplicantsIncome_count = len(loan.ApplicantIncome)

Best_ApplicantsIncome = [ApplicantIncome for ApplicantIncome in loan.ApplicantIncome if not math.isnan(ApplicantIncome)]
Best_ApplicantsIncome_count = len(Best_ApplicantsIncome)
mean_ApplicantsIncome = sum(Best_ApplicantsIncome)/Best_ApplicantsIncome_count

print(f'count of all ApplicantIncome = {ApplicantsIncome_count}')
print(f'count of Best ApplicantIncome = {Best_ApplicantsIncome_count}')
print(f'mean of Best ApplicantIncome = {mean_ApplicantsIncome:.2f}')
```

# Let's calculate Measures of Variablity (Range and Percentile) from the numpy library 

```python
import numpy as np

def ApplicantIncome_range(Best_ApplicantsIncome):
    """
    Compute the range of a list or array of values.
    @param values the list or array of values.
    @return the range
    """

    return np.amax(Best_ApplicantsIncome) - np.amin(Best_ApplicantsIncome)


import numpy as np

np_ApplicantIncome_range = ApplicantIncome_range(Best_ApplicantsIncome)

print(f'np.amin(Best_ApplicantsIncome) = {np.amin(Best_ApplicantsIncome)}')
print(f'np.amax(Best_ApplicantsIncome) = {np.amax(Best_ApplicantsIncome)}')
print(f'Income_range(Best_ApplicantsIncome) = {np_ApplicantIncome_range}')

##"Calculating Percentiles First Quartile as q1, second quartile as q2 and third quartile as q3"
##""

q3 = np.percentile(Best_ApplicantsIncome, 75)
q2 = np.percentile(Best_ApplicantsIncome, 50)
q1 = np.percentile(Best_ApplicantsIncome, 25)

print(f'q3 = {q3}')
print(f'q1 = {q1}')
## formula to calculate Interquertile range of loan applicants Income

print(f'q3 - q1 = {q3 - q1}') 
```

```python
def Income_and_propertyarea(propertyareatype):
    """ 
    Return the list of incomes of home loan applicant having particular property area .
    @param the given property area type.
    @return the list of incomes.
    """
    # Count only best appliants income.
    return [ApplicantIncome for ApplicantIncome, pat in zip(loan.ApplicantIncome, loan.Property_Area) 
                         if not math.isnan(ApplicantIncome) and (pat == propertyareatype)]
```

```python
import seaborn as sns

Propertytype = ['Urban', 'Semiurban', 'Rural']
Applicantsincomecounts  = [AI_U, AI_SU, AI_R]

sns.set_style('whitegrid')  # white backround with gray grid lines
axes = sns.barplot(Propertytype, Applicantsincomecounts, palette='bright')
axes.set_title('Count of applicantincomes with each property type')
axes.set(xlabel='Applicant Property type', ylabel='Applicants Income')

# Scale the y-axis by 10% to make room for text above the bars.
axes.set_ylim(top=1.10*max(Applicantsincomecounts))

# Display the count above each patch (bar).
for bar, count in zip(axes.patches, Applicantsincomecounts):
    text_x = bar.get_x() + bar.get_width()/2  
    text_y = bar.get_height() 
    text   = f'{count}'
    axes.text(text_x, text_y, text, 
              fontsize=11, ha='center', va='bottom')

plt.show()  # display the chart 
```
