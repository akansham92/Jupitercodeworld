```python
"""Importing libraries required"""
import random
import numpy as np
import seaborn as sns
```

```python

"""Importing libraries required for plotting Histogram"""

%matplotlib inline
import matplotlib.pyplot as plt


"""Declaring a variable name Exponential_Values to assign lambda value as 12.5"""
Exponential_Values = []
lambd = 1/12.5
while len(Exponential_Values) < 1000:
    x = random.expovariate(lambd)
    if x < 2*25:
        Exponential_Values += [x]
    
"""Histogram X and Y axis data"""    
Value, Frequency = np.unique(Exponential_Values, return_counts=True)

plt.figure(figsize=(10,6))
plt.hist([random.expovariate(lambd) for i in range(1000)], bins = 150)
sns.set_style('whitegrid')
axes= sns.histplot(Exponential_Values)
axes.set(xlabel='Value', ylabel = 'Frequency', title='Exponential Distribution of 1000 Values')






```
