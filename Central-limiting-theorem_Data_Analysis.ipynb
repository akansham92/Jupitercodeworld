To conduct an experiment on the Central Limit Theorem, I am using Home loan sanction dataset which consists of amount of the ApplicantIncome, CoapplicantIncome, LoanAmount and Property_Area for 365 Home loan applicants.

I will start by calculating the parameters like mean, median, standard deviation and interquartile range (IQR) for the ApplicantIncome feature.

```python
"""Importing libraries required"""
import pandas as pd
import random as rdm
import numpy as np
import seaborn as sns


# Load the dataset into a dataframe.
Pop_loan = pd.read_csv('/Users/Checkout/220Assignment/Week6/loan_sanction_test.csv')

# calculate population parameters
ApplicantIncome_mean = Pop_loan['ApplicantIncome'].mean()
ApplicantIncome_median = Pop_loan['ApplicantIncome'].median()
ApplicantIncome_std = Pop_loan['ApplicantIncome'].std()
ApplicantIncome_iqr = Pop_loan['ApplicantIncome'].quantile(0.75) - Pop_loan['ApplicantIncome'].quantile(0.25)

print("Population Mean of ApplicantIncome:", ApplicantIncome_mean)
print("Population Median of ApplicantIncome:", ApplicantIncome_median)
print("Population STD of ApplicantIncome:", ApplicantIncome_std)
print("Population IQR of ApplicantIncome:", ApplicantIncome_iqr)
```

```python


"""Declaring a variable name Normal_Values to store 365 integers"""
Normal_Values = [rdm.gauss(4805.6,6) for i in range(365)]

"""Histogram X and Y axis data"""
Value, Frequency = np.unique(Normal_Values, return_counts=True)

"""Importing libraries required for plotting Histogram"""
%matplotlib inline
import matplotlib.pyplot as plt


"""We can customise the size,title,Xlabel and Y label as per requirements from here"""

plt.figure(figsize=(8,4))
sns.set_style('whitegrid')

axes= sns.histplot(Normal_Values, kde=True)
axes.set(xlabel='ApplicantIncomeValues', ylabel = 'Frequency', title='Normal Distribution of 365 loan applicants');
```

Next, I will be performing random sampling to estimate the random sampling of the population parameters.
I will use different sample numbers to see how the sample statistics converge to the population parameters as predicted by the Central Limit Theorem.

```python
# Perform random sampling and calculate sample statistics
# Experiment 1: 10 samples of size 10
exp1_means = []
exp1_medians = []
exp1_iqrs = []
for i in range(1000):
    sample = Pop_loan['ApplicantIncome'].sample(n=10, replace=True)
    exp1_means.append(sample.mean())
    exp1_medians.append(sample.median())
    exp1_iqrs.append(np.percentile(sample, 75) - np.percentile(sample, 25))
```

```python
# Experiment 2: 100 samples of size 30
exp2_means = []
exp2_medians = []
exp2_iqrs = []
for i in range(1000):
    sample = Pop_loan['ApplicantIncome'].sample(n=30, replace=True)
    exp2_means.append(sample.mean())
    exp2_medians.append(sample.median())
    exp2_iqrs.append(np.percentile(sample, 75) - np.percentile(sample, 25))
```

```python
# Experiment 3: 1000 samples of size 100
exp3_means = []
exp3_medians = []
exp3_iqrs = []
for i in range(1000):
    sample = Pop_loan['ApplicantIncome'].sample(n=100, replace=True)
    exp3_means.append(sample.mean())
    exp3_medians.append(sample.median())
    exp3_iqrs.append(np.percentile(sample, 75) - np.percentile(sample, 25))
```

```python
#Create matplotlib charts to illustrate the results:
# Create histograms of sample means
fig, axs = plt.subplots(1, 3, figsize=(12, 4))
sns.histplot(exp1_means, kde=True, ax=axs[0])
sns.histplot(exp2_means, kde=True, ax=axs[1])
sns.histplot(exp3_means, kde=True, ax=axs[2])
axs[0].set_title("Experiment 1")
axs[1].set_title("Experiment 2")
axs[2].set_title("Experiment 3")
plt.suptitle("Histograms of Sample Means")
plt.tight_layout()
plt.show()

# Create histograms of sample medians
fig, axs = plt.subplots(1, 3, figsize=(12, 4))
sns.histplot(exp1_medians, kde=True, ax=axs[0])
sns.histplot(exp2_medians, kde=True, ax=axs[1])
sns.histplot(exp3_means, kde=True, ax=axs[2])
axs[0].set_title("Experiment 1")
axs[1].set_title("Experiment 2")
axs[2].set_title("Experiment 3")
plt.suptitle("Histograms of Sample Medians")
plt.tight_layout()
plt.show()
```
