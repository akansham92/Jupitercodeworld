```python
"""Importing libraries required"""

import random as rdm
import numpy as np
import seaborn as sns

```

```python
"""Declaring a variable name Normal_Values to store 1000 integers"""
Normal_Values = [rdm.gauss(12.5,4) for i in range(1000)]

"""Histogram X and Y axis data"""
Value, Frequency = np.unique(Normal_Values, return_counts=True)
```

```python

"""Importing libraries required for plotting Histogram"""
%matplotlib inline
import matplotlib.pyplot as plt


"""We can customise the size,title,Xlabel and Y label as per requirements from here"""

plt.figure(figsize=(10,6))
sns.set_style('whitegrid')

axes= sns.histplot(Normal_Values)
axes.set(xlabel='Value' , ylabel = 'Frequency', title='Normal Distribution of 1000 Values');
```
